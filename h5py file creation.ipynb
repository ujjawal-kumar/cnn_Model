```python

hf = h5py.File('data.h5', 'w')

for i in l1:
    list1=[]
    path1=os.path.join(path,i)
    l2=os.listdir(path1)
    for j in l2:
        path2=os.path.join(path1,j)
        img1=imread(path2)
        if len(img1.shape) == 2: # if img is grayscale, expand
            #print ("convert 1-channel image to 3-channel image.")
            new_img = np.zeros((img1.shape[0],img1.shape[1],3))
            for ch in range(3):
                for xx in range(img1.shape[0]):
                    for yy in range(img1.shape[1]):
                        new_img[xx,yy,ch] = img1[xx,yy]
            img1= new_img
        img2=imresize(img1,(224,224))
        list1.append(img2)
        
    hf.create_dataset(str(i), data=list1, compression="gzip", compression_opts=9)
    
plt.imshow(img2)
plt.show()
hf.close()
    
```
